DTU PLACEMENT DATA 2024-2025 BATCH 


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
```

```python
df = pd.read_excel('placement.xlsx', sheet_name='Sheet1')
print(df.head())
```

```python
print(df.columns)
```

```python
df = df.drop(columns=['Details', '#', 'Additional Info'])
```

```python
df['Role'] = df['Role'].fillna('NAN')
df['CTC'] = df['CTC'].fillna('NAN')
df['Company'] = df['Company'].fillna('NAN')
df['Stpd'] = df['Stpd'].fillna('NAN')
```

```python
df['CGPA'] = pd.to_numeric(df['CGPA'], errors = 'coerce')
df['CTC'] = pd.to_numeric(df['CTC'], errors='coerce')
df['Stpd'] = pd.to_numeric(df['Stpd'], errors='coerce')
```

```python
# Define the CTC ranges (bins)
bins = [0, 10, 20, 50, 100]  # Define ranges in LPA
labels = ['0-10', '10-20', '20-50', '50-100']  # Labels for the ranges

# Categorize the CTC values into the defined bins
df['CTC Range'] = pd.cut(df['CTC'], bins=bins, labels=labels, right=False)

# Count the number of students in each range
ctc_range_counts = df['CTC Range'].value_counts().sort_index()

# Plot the distribution
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=ctc_range_counts.index, y=ctc_range_counts.values, palette='Set2')

for i, count in enumerate(ctc_range_counts.values):
    ax.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=10)


plt.title('Number of Students in Different CTC Ranges')
plt.xlabel('CTC Range (LPA)')
plt.ylabel('Number of Students')
plt.show()
```
